# Principal Axis Transformation

<span style='font-size:medium'>Janina Planeta, Julia Renner</span>  



<span style='font-size:medium'>The principal axis theorem states that the principal axes of quadrics are perpendicular and gives a procedure to find them.</span>

<span style='font-size:medium'>The algorithm consists of transforming coordinates of a given equation. The main tool is the diagonalization of symmetric matrices through the multiplication with orthogonal matrices. Visually, this translates to rotations, translations and reflections.</span>

<span style='font-size:medium'>Through this method we can determine the type of a given quadric and its geometrical characteristics, such as center, axes,</span><span style='font-size:medium'><span style='color:#f44336'> </span></span><span style='font-size:medium'>vertex,...</span>  



### Reminder



#### Quadric

<span style='font-size:medium'>A quadric is a point set in </span>$\mathbb {R} ^{n}$<span style='font-size:medium'> which satisfies  </span>$Q=\left\{(x_{1},\ldots ,x_{n})\in \mathbb {R} ^{n}\mid q(x_{1},\ldots ,x_{n})=0\right\}$<span style='font-size:medium'> where </span>$q(x_{1},\ldots ,x_{n})=\sum _{i,j=1}^{n}a_{ij}x_{i}x_{j}+2\,\sum _{i=1}^{n}b_{i}x_{i}+c$<span style='font-size:medium'>.</span>  



$Q_e = \{x \in \mathbb{R}^2 | \frac{x_1^2}{\alpha^2} + \frac{x_2^2}{\beta^2} = 1 \}$  



 $Q_h = \{x \in \mathbb{R}^2 |  \frac{x_1^2}{\alpha^2} - \frac{x_2^2}{\beta^2} = 1 \}$  



$Q_p = \{x \in \mathbb{R}^2 |  \frac{x_1^2}{\alpha^2} - 2 x_2 = 0 \}$



#### Characteristics of matrices

<span style='font-size:medium'>**Symmetric matrix:**</span><span style='font-size:medium'> </span>$A=A^T$

<span style='font-size:medium'>**Orthogonal matrix:**</span><span style='font-size:medium'> </span>$A \cdot A^T =Id$<span style='font-size:medium'>  </span>  



### Method



<span style='font-size:medium'>For this presentation we will only look at the two\-dimensional case. Visually, we will first rotate and then move the coordinate system in which our object is located.</span>  



<img src=".Principal axis transformation (3) (1).ipynb.upload/paste-0.20758429196338946"   width="591.98px"  height="174.82px"  style="object-fit:cover"/>  



<span style='font-size:medium'>We get an equation of the form </span>$ax^2 + bxy +cy^2 + dx + ey + f = 0  \ \ (a,b,c,d,e,f \in \mathbb{R})$<span style='font-size:medium'> and we want to bring it in the form </span>

$Q = \{x \in \mathbb{R}^2 | \frac{x_1^2}{\alpha^2} \pm \frac{x_2^2}{\beta^2} = 1 \}$<span style='font-size:medium'> or </span>$Q = \{x \in \mathbb{R}^2 |  \frac{x_1^2}{\alpha^2} - 2 x_2 = 0 \}$<span style='font-size:medium'>.  Let  </span>$v := \left(\begin{matrix} x \\ y\end{matrix}\right)$. <span style='font-size:medium'>We write the equation in matrix form, that is</span>

$v^T\left( \begin{matrix} a & b/2 \\ b/2 & c \end{matrix} \right) v + \left(\begin{array}{c} d & e \end{array}\right)v +f = 0$.  



```python
# show("a = ")
a = float(input())
show("b = ")
b = float(input())
show("c = ")
c = float(input())
show("d = ")
d = float(input())
show("e = ")
e = float(input())
show("f = ")
f = float(input())
```

```python
coeff = [a, b, c, d, e, f]
pos_neg = ["+", "+", "+", "+", "+", "+"]
i = 0
for k in coeff:
    if sgn(k) == 1:
        pos_neg[i] = "+"
    else: 
        pos_neg[i] = "-"
    i = i+1
if pos_neg[0] == "+":
    pos_neg[0] = ""
    
abs_coeff = [abs(k) for k in coeff]

tex(f" $$ {latex(pos_neg[0])}{latex(abs_coeff[0])} x^2 {latex(pos_neg[1])}{latex(abs_coeff[1])} xy {latex(pos_neg[2])}{latex(abs_coeff[2])} y^2 {latex(pos_neg[3])} {latex(abs_coeff[3])} x {latex(pos_neg[4])}{latex(abs_coeff[4])} y {latex(pos_neg[5])} {latex(abs_coeff[5])} = 0 $$ ")
```

<span style='font-size:medium'>Geometrically, this equation looks like this.</span>  



```python
var('x,y')
xmin = -6
xmax = 6
ymin = -6
ymax = 6
implicit_plot(a*x^2 + b*x*y+ c*y^2 + e*y + f==0, (x,xmin,xmax), (y, ymin,ymax)) + line([(xmin,0),(xmax,0)], rgbcolor = (0,0,0)) + line([(0, ymin),(0,ymax)], rgbcolor = (0,0,0))
```

#### <span style='font-size:medium'>**Step 1: Determine the matrix from the equation in**</span><span style='font-size:medium'> </span><span style='font-size:medium'>**matrix form**</span>

$A = \left( \begin{matrix} a & b/2 \\ b/2 & c \end{matrix} \right)$   



```python
A = matrix(QQ, [[a, b/2], [b/2, c]])
show(A)
```

```python
z = column_matrix(QQ, [d, e]).transpose()

tex(f" $$ v^t {latex(A)} v + {latex(z)} v {latex(pos_neg[5])} {latex(abs_coeff[5])} = 0 $$ ")
```

#### <span style='font-size:medium'>**Step 2: Determine the eigenvalues**</span><span style='font-size:medium'> </span>$\lambda_1$<span style='font-size:medium'> **and** </span>$\lambda_2$



```python
ev_A = A.eigenvalues()
show(ev_A)
```

<span style='font-size:medium'>**Step 3: Find the eigenvectors and normalize them**</span>

<span style='font-size:medium'>When we have found the eigenvectors </span>$v_1$<span style='font-size:medium'> and </span>$v_2$<span style='font-size:medium'>, we also know the principal axes. In the two\-dimensional case these are given by </span>$\mathbb{R}v_1$<span style='font-size:medium'>and </span>$\mathbb{R}v_2$<span style='font-size:medium'>.</span>  



```python
# find the eigenvectors
I = identity_matrix(2)
eigenvectors = []

for k in ev_A:
    eigenvectors.append((A - k*I).right_kernel().basis())

evs = eigenvectors

# if they are the same, save only one of them
if ev_A[0] == ev_A[1]:
    evs.pop()
    
show(evs)
```

```python
# normalize the eigenvectors
for k in range(0, len(evs)):
    v = list(evs[k][0])
    norm_squared = v[0]**2 + v[1]**2
    norm = sqrt(norm_squared)
    v[0] = v[0] * (1/norm)
    v[1] = v[1] * (1/norm)
    evs[k] = tuple(v)
    
show(evs)
```

<span style='font-size:medium'>**Step 4: Create a new matrix S whose columns correspond to the eigenvectors**</span>  

<span style='font-size:medium'>A is a symmetric matrix in </span>$\mathbb{R}$<span style='font-size:medium'>, which implies that it is diagonalizable. </span>$\Rightarrow{} A = S DS^T$<span style='font-size:medium'> where </span>$D = \left( \begin{matrix}\lambda_1 & 0 \\ 0 &  \lambda_2 \end{matrix} \right)$<span style='font-size:medium'>and </span>$S$<span style='font-size:medium'> is orthogonal.</span>  



```python
S = matrix([evs[0], evs[1]])
show(S)
```

<span style='font-size:medium'>**Step 5: Getting rid of mixed terms by substitution \(rotation\)**</span>

<span style='font-size:medium'>We substitute </span>$A = SDS^T$<span style='font-size:medium'> and add </span>$SS^T \ (=Id$<span style='font-size:medium'>, because </span>$S$<span style='font-size:medium'> is orthogonal </span>$)$<span style='font-size:medium'> in </span>$v^T A v + \left(\begin{array}{c} d & e \end{array}\right) v +f = 0$<span style='font-size:medium'>.</span>  

$\Rightarrow v^T SDS^T v + \left(\begin{array}{c} d & e \end{array}\right) SS^Tv +f = 0$

<span style='font-size:medium'>Write </span>$v$<span style='font-size:medium'> in dependence of </span>$S$<span style='font-size:medium'> and a new vector </span>$w :=\left( \begin{array}{c} s \\ t \end{array} \right)$<span style='font-size:medium'>: </span>$v = S w$ . 

$\Rightarrow w^TS^T SDS^T Sw + \left(\begin{array}{c} d & e \end{array}\right) SS^T Sw +f = 0$

$\Rightarrow w^TDw + \left(\begin{array}{c} d & e \end{array}\right) Sw +f = 0$

<span style='font-size:medium'>The new equation is now </span>$\lambda_1 \cdot s^2 + \lambda_2 \cdot t^2 + g \cdot s +h \cdot t + f = 0 \ \ g,h$<span style='font-size:medium'>  are new constants resulting from multiplication by S.</span>  



```python
tex(f" $$ v^t {latex(S)}{latex(A)}{latex(S)}^T v + {latex(z)} {latex(S)} {latex(S)}^T v {latex(pos_neg[5])} {latex(abs_coeff[5])}  = 0 $$ ")
```

```python
D = S*A*S.transpose()

z_new = z*S

tex(f" $$ w^t {latex(D)} w + {latex(z_new)} w {latex(pos_neg[5])} {latex(abs_coeff[5])}  = 0 $$ ")
```

```python
#new equation
new_coeff = [ev_A[0], ev_A[1], z_new[0][0], z_new[0][1], coeff[5]]
pos_neg_new = ["+", "+", "+", "+", "+"]
i = 0
for k in new_coeff:
    if sgn(k) == 1:
        pos_neg_new[i] = "+"
    else: 
        pos_neg_new[i] = "-"
    i = i+1
if pos_neg_new[0] == "+":
    pos_neg_new[0] = ""
    
abs_new_coeff = [abs(k) for k in new_coeff]

tex(f" $$ {latex(pos_neg_new[0])}{latex(abs_new_coeff[0])} s^2 {latex(pos_neg_new[1])}{latex(abs_new_coeff[1])} t^2 {latex(pos_neg_new[2])}{latex(abs_new_coeff[2])} s {latex(pos_neg_new[3])}{latex(abs_new_coeff[3])} t {latex(pos_neg_new[4])}{latex(abs_new_coeff[4])} = 0 $$ ")
```

<span style='font-size:medium'>Visually, we rotate the coordinate system so that the symmetry axes of the object are parallel to the new coordinate axes afterwards. The normalized eigenvectors are the basis vectors of the new coordinate system.</span>  



```python
var('s,t')
implicit_plot(new_coeff[0]*s^2 + new_coeff[1]*t^2 + new_coeff[2]*s + new_coeff[3]*t + new_coeff[4]==0, (s,xmin,xmax), (t, ymin,ymax)) + line([(xmin,0),(xmax,0)], rgbcolor = (0,0,0)) + line([(0, ymin),(0,ymax)], rgbcolor = (0,0,0))
```

<span style='font-size:medium'>If either </span>$\lambda_1$<span style='font-size:medium'> or </span>$\lambda_2$<span style='font-size:medium'> is zero, i.e. the object is a parabola, we do the same thing as in the following with only one variable.</span>  



<span style='font-size:medium'>**Step 6: Getting rid of the non\-square terms \(shift\)**</span>

<span style='font-size:medium'>The end goal is to get an equation of the form </span>$\frac{(s-\xi)^2}{\alpha^2} + \frac{(t- \eta)^2}{\beta^2} = 1$<span style='font-size:medium'>. Whereas </span>$(\xi,\ \eta )$<span style='font-size:medium'> is the center of our object in the original coordinate system and </span>$\alpha$<span style='font-size:medium'> and </span>$\beta$<span style='font-size:medium'> are the half\-axes.</span>  

$\lambda_1 \cdot s^2 + \lambda_2 \cdot t^2 + g \cdot s +h \cdot t + f = 0$

$\Rightarrow \lambda_1 \cdot (s^2 + \frac{g}{\lambda_1} \cdot s + (\frac{g}{2\lambda_1})^2 - (\frac{g}{2\lambda_1})^2) + \lambda_2 \cdot ( t^2 + \frac{h}{\lambda_2} \cdot t + (\frac{h}{2\lambda_2})^2 - (\frac{h}{2\lambda_2})^2) =-f$

$\Rightarrow \lambda_1 \cdot (s+ \frac{g}{2\lambda_1})^2 + \lambda_2 \cdot (t+ \frac{h}{2\lambda_2})^2 = -f + \frac{g^2}{4\lambda_1} + \frac{h^2}{4\lambda_2}$

$\Rightarrow \frac{(s-\xi)^2}{\alpha^2} + \frac{(t- \eta)^2}{\beta^2} = 1$  



```python
la = []

# we have to check that the eigenvalues are not equal to zero else we get problems with division
for k in range(0, len(ev_A)):
    if ev_A[k] == 0:
        la.append(1)
    else:
        la.append(ev_A[k])
        

# find center
list(z_new)

m = z_new[0][0]/(2*la[0])
n = z_new[0][1]/(2*la[1])

center = [-m, -n]

# we substract what we added to the equation
j = float(-coeff[5] + ev_A[0]*(m**2) + ev_A[1]*(n**2))
        
show(center)

show(j)
```

```python
# plot if both of the eigenvalues are not zero
alpha = sqrt(ev_A[0]/j)
beta = sqrt(ev_A[1]/j)

pos_neg_ev_A = ["+", "+"]
i = 0
for k in ev_A:
    if sgn(k) == 1:
        pos_neg_ev_A[i] = "+"
    else: 
        pos_neg_ev_A[i] = "-"
    i = i+1
if pos_neg_ev_A[0] == "+":
    pos_neg_ev_A[0] = ""
    
pos_neg_center = ["+", "+"]
i = 0
for k in center:
    if sgn(k) == 1:
        pos_neg_center[i] = "-"
    else: 
        pos_neg_center[i] = "+"
    i = i+1

abs_center = [abs(k) for k in center]

tex(f" $$ {latex(pos_neg_ev_A[0])} (s {latex(pos_neg_center[0])} {latex(abs_center[0])})^2 /({latex(alpha)})^2 {latex(pos_neg_ev_A[1])} (t {latex(pos_neg_center[1])}{latex(abs_center[1])})^2 /({latex(beta)})^2 = 1$$ ")
```

```python
tex(f" $$ {latex(pos_neg_ev_A[0])} o^2 /({latex(alpha)})^2 {latex(pos_neg_ev_A[1])} p^2 /({latex(beta)})^2 = 1$$ ")
```

<span style='font-size:medium'>Visually, we shift the coordinate system so that the center of the object is in zero.</span>  



```python
var('o,p')
implicit_plot(ev_A[0]/j*o^2 + ev_A[1]/j*p^2 == 1, (o,xmin,xmax), (p, ymin,ymax)) + line([(xmin,0),(xmax,0)], rgbcolor = (0,0,0)) + line([(0, ymin),(0,ymax)], rgbcolor = (0,0,0))
```
