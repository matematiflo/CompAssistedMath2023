In this project, we were asked to write a `Sage` program that:

1. Asks the user to input an integer $ n $ and family of vectors in the vector space $ E = \mathbb{Q}^n $.
1. Extracts from this family a basis of the subspace $ F $ generated by these vectors.
1. Complete this basis to a basis of $ \mathbb{ Q }^n $ and extract from it a basis of an explicit complement of $ F $ in $ \mathbb{ Q }^n $.



First we ask the user to input a number $n$ that represents the dimension of the vector subspace we will examine.



```python
#input dimension
n=int(input("Please Enter the dimension of your vector space: "))
```

Then we construct a loop that asks the user to input $n$ vectors of dimension $n$.


```python
#input vectors
vectors=[]
for i in range(n):
    v=tuple(map(int,input("Please Enter {} vectors of dimension {}, separating the coefficients by spaces".format(n,n)).split()))
    vectors.append(vector(QQ,v))

```

The function `span()` determines the span of the vectors given by the users. The `show()`function outputs the span.



```python
V=span(vectors)
#Output span         
show(V)
```

With `.basis()` we create a basis of the span and output it with `show()`.



```python
#2. extracting the basis
B=V.basis()
show(B)
```

To complete our basis we write a for loop that creates $n$-dimensional unit vectors. We then check wether these vectors are already included in the span of our user created vectors and delete them if they are. If they are not we add them to the basis of the complement with the `.append()` function.



```python
#3. completing the basis and constructing the complement
complement_basis = []
for i in range(n):
    e = vector(QQ, [0]*n)
    e[i] = 1
    if e not in V:
        complement_basis.append(e)
```

Finally, we use an if-else statement to determine wether the complement of our subspace is empty and output the answer and the basis of the complement if it s non empty.



```python
if complement_basis:
    C = span(complement_basis)
    show("A basis for the complement of F is:", complement_basis)
else:
    print("The complement of F is empty.")
show("A basis for F is:", B)
```
